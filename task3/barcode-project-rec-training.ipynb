## Prepare localization model

```python
MODEL = 'COCO-Keypoints/keypoint_rcnn_R_101_FPN_3x.yaml'
```

```python
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file(MODEL))
cfg.MODEL.WEIGHTS = "/kaggle/input/barcodelocmodel/model_final.pth"
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 512
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1
cfg.MODEL.ROI_KEYPOINT_HEAD.NUM_KEYPOINTS = 4
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.9
predictor = DefaultPredictor(cfg)
```

```python
def get_points(preds, s=15):
    top_left_x = min(preds[:, 0])
    top_left_y = min(preds[:, 1])
    bot_right_x = max(preds[:, 0])
    bot_right_y = max(preds[:, 1])
    
    last_points = []
    for x, y in preds:
        if (x == top_left_x and y == top_left_y) or (x == bot_right_x and y == bot_right_y):
            continue
        last_points.append([x, y])
    last_points = np.array(last_points)
    
    top_right_x = max(last_points[:, 0])
    top_right_y = min(last_points[:, 1])
    bot_left_x = min(last_points[:, 0])
    bot_left_y = max(last_points[:, 1])
    
    return top_left_x - s, top_left_y - s, bot_left_x - s, bot_left_y + s, bot_right_x + s, bot_right_y + s, top_right_x + s, top_right_y - s
```

```python
from tqdm import tqdm

for root, dirname, filenames in os.walk('/kaggle/input/barcode-dataset/Train/Train/Images'):
    for filename in tqdm(filenames):
        if filename not in imgs_annotated:
            continue
        
        im_path = os.path.join(root, filename)
        
        points = []
        values = annotations.loc[annotations['img'] == filename, 'x1':'y4'].values[0]
        for i in range(0, 8, 2):
            points.append([values[i], values[i + 1]])
        points = np.array(points)
        
        im = Image.open(im_path).convert('RGB')
        width, height = im.size
        points = get_points(points, s=width//12)
        result = im.transform((width, height), ImageTransform.QuadTransform(points))

        result.save(f'/kaggle/barcodes/{filename}')
```
